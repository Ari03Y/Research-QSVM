```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import OneClassSVM
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from sklearn.model_selection import train_test_split
from tabulate import tabulate

# Qiskit imports
from qiskit.circuit.library import ZZFeatureMap, PauliFeatureMap
from qiskit_machine_learning.kernels import FidelityQuantumKernel

# Set random seed for reproducibility
np.random.seed(42)

# ---------------------------
# Data Loading and Preparation
# ---------------------------

# Assume 'data_noisy' is already loaded as a pandas DataFrame
# Example:
# data_noisy = pd.read_csv('your_data.csv')

# Display data shape and first few rows
print("Data shape:", data_noisy.shape)
print("First few rows:\n", data_noisy.head())

# ---------------------------
# Data Preprocessing
# ---------------------------

# Standard scaler
scaler = StandardScaler()

# Apply scaling
data_scaled = scaler.fit_transform(data_noisy)

# Dimensionality Reduction using PCA
max_qubits = 4  # Increased from 2 to 4 for better feature representation
pca = PCA(n_components=max_qubits)
data_reduced = pca.fit_transform(data_scaled)

# Convert back to DataFrame
data_reduced = pd.DataFrame(data_reduced, columns=[f'PC{i+1}' for i in range(max_qubits)])

# ---------------------------
# Define Applications and Prepare Results Dictionary
# ---------------------------

# Applications dictionary
applications = {
    'blackscholes': 22,
    'bodytrack': 18,
    'canneal': 18,
    'x264': 17,
    'dedup': 16,
    'ferret': 16,
    'fluidanimate': 16,
    'raytrace': 16,
    'streamcluster': 16,
    'swaptions': 16,
    'vips': 16,
    'freqmine': 15
}

# Prepare results dictionary
results = {
    app.capitalize(): {
        'QSVM': {'Precision': 0, 'Recall': 0, 'F1-score': 0, 'Accuracy': 0}
    } for app in applications
}

# ---------------------------
# Define Quantum Feature Map and Kernel
# ---------------------------

# Define the feature map (you can switch to PauliFeatureMap if desired)
feature_map = ZZFeatureMap(feature_dimension=max_qubits, reps=1, entanglement='linear')  # reps=1 for speed

# Initialize the FidelityQuantumKernel
quantum_kernel = FidelityQuantumKernel(feature_map=feature_map)

# Define kernel function compatible with scikit-learn
def kernel_func(X, Y):
    return quantum_kernel.evaluate(X, Y)

# ---------------------------
# Train and Evaluate QSVM with Manual Hyperparameter Tuning
# ---------------------------

# Define hyperparameter grid
nu_values = [0.05, 0.1, 0.15, 0.2, 0.25]

for app, rows in applications.items():
    print(f"Processing application: {app.capitalize()}")
    # Select the subset of data for the current application
    subset_data = data_reduced.iloc[:rows]
    
    # Split into training and testing sets
    # For One-Class SVM, training is done on inliers only
    # Here, we'll simulate outliers by generating synthetic outliers
    train_data, test_data_inliers = train_test_split(subset_data, test_size=0.3, random_state=42)
    
    # Generate synthetic outliers
    num_outliers = int(0.3 * rows)
    outliers = np.random.uniform(low=-4, high=4, size=(num_outliers, max_qubits))
    outliers = pd.DataFrame(outliers, columns=[f'PC{i+1}' for i in range(max_qubits)])
    outliers['Label'] = 0  # Label outliers as 0
    
    # Prepare test data with inliers and outliers
    test_data = test_data_inliers.copy()
    test_data['Label'] = 1  # Label inliers as 1
    outliers['Label'] = 0    # Label outliers as 0
    test_data = pd.concat([test_data, outliers], ignore_index=True)
    
    # Separate features and labels
    X_train = train_data.values
    X_test = test_data.drop('Label', axis=1).values
    y_test = test_data['Label'].values
    
    best_f1 = -1
    best_nu = None
    best_metrics = {'Precision': 0, 'Recall': 0, 'F1-score': 0, 'Accuracy': 0}
    
    for nu in nu_values:
        # Define the One-Class SVM with QSVM kernel
        svm = OneClassSVM(kernel=kernel_func, nu=nu)
        
        try:
            # Fit the model on training data
            svm.fit(X_train)
            
            # Predict on test data
            pseudo_labels = svm.predict(X_test)
            # Convert labels: +1 to 1 (inlier), -1 to 0 (outlier)
            predicted_labels = (pseudo_labels == 1).astype(int)
            
            # Calculate performance metrics
            precision = precision_score(y_test, predicted_labels, zero_division=0)
            recall = recall_score(y_test, predicted_labels, zero_division=0)
            f1 = f1_score(y_test, predicted_labels, zero_division=0)
            accuracy = accuracy_score(y_test, predicted_labels)
            
            # Update best parameters based on F1-score
            if f1 > best_f1:
                best_f1 = f1
                best_nu = nu
                best_metrics = {
                    'Precision': precision,
                    'Recall': recall,
                    'F1-score': f1,
                    'Accuracy': accuracy
                }
        except Exception as e:
            print(f"An error occurred with nu={nu} for application {app.capitalize()}: {e}")
            continue
    
    # Store the best results
    results[app.capitalize()]['QSVM'] = best_metrics
    print(f"Best nu for {app.capitalize()}: {best_nu}")
    print(f"Precision: {best_metrics['Precision']}, Recall: {best_metrics['Recall']}, F1-score: {best_metrics['F1-score']}, Accuracy: {best_metrics['Accuracy']}\n")

# ---------------------------
# Present Results
# ---------------------------

# Prepare tabulated results
tabulate_data = []
for app, kernels_data in results.items():
    qsvm_data = kernels_data['QSVM']
    row = [
        app,
        f'{qsvm_data["Accuracy"]:.2f}' if not np.isnan(qsvm_data["Accuracy"]) else 'NaN',
        f'{qsvm_data["Precision"]:.2f}' if not np.isnan(qsvm_data["Precision"]) else 'NaN',
        f'{qsvm_data["Recall"]:.2f}' if not np.isnan(qsvm_data["Recall"]) else 'NaN',
        f'{qsvm_data["F1-score"]:.2f}' if not np.isnan(qsvm_data["F1-score"]) else 'NaN'
    ]
    tabulate_data.append(row)

headers = ['Application', 'QSVM Accuracy', 'QSVM Precision', 'QSVM Recall', 'QSVM F1-score']
print(tabulate(tabulate_data, headers=headers, tablefmt='grid'))
```
